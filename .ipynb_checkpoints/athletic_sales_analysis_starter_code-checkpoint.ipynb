```python
# Import Libraries and Dependencies
import pandas as pd
import numpy as np
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020 = pd.read_csv("Resources/athletic_sales_2020.csv")
sales_2021 = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
sales_2020
```

```python
# Display the 2021 sales DataFrame
sales_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
print(sales_2020.dtypes)
```

```python
# Check the 2021 sales data types.
print(sales_2021.dtypes)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([sales_2020, sales_2021], axis = 0)
combined_sales_df = combined_sales_df.reset_index(drop=True)
combined_sales_df
```

```python
# Check if any values are null.
combined_sales_df.isnull().sum()
```

```python
# Check the data type of each column
print(combined_sales_df.dtypes)
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_df['invoice_date'] = pd.to_datetime(combined_sales_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(combined_sales_df.dtypes)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.

products_sold_df = combined_sales_df.groupby(['region', 'state', 'city']).size().reset_index()

# Rename the sum to "Total_Products_Sold".

Total_Products_Sold = products_sold_df.rename(columns={'products_sold_df': 'Total_Products_Sold'})

# Show the top 5 results.

Total_Products_Sold.head()
                            
```

#### Using `pivot_table`

```python
units_sold_df = combined_sales_df.pivot_table(values='product', index=['region', 'state', 'city'], aggfunc='sum')
Total_Products_Sold = units_sold_df.rename(columns={'product': 'Total_Products_Sold'})
Total_Products_Sold.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_df = combined_sales_df.groupby(["region", "state", "city"])[["total_sales"]].sum()

# Rename the "total_sales" column to "Total Sales"

total_sales = total_sales_df.rename(columns={'total_sales': 'Total_Sales'})

# Show the top 5 results.
total_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

total_sales = pd.pivot_table(combined_sales_df, values='total_sales', index=['region', 'state', 'city'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
total_sales.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales_data = combined_sales_df[combined_sales_df['total_sales'] == 'Women\'s Athletic Footwear']

```

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

units_sold_df = womens_footwear_sales_data.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
units_sold_df = units_sold_df.rename(columns={'qty': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
units_sold_df.head()
```

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table = pd.pivot_table(combined_sales_df, index='invoice_date', values='total_sales')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table.rename(columns={'total_sales': 'Total Sales'}, inplace=True)


# Show the table.
pivot_table
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = pivot_table.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=True)

sorted_daily_sales
```

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = pivot_table.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
sorted_weekly_sales = weekly_sales.sort_values(by='Total Sales', ascending=True)

sorted_weekly_sales
```
